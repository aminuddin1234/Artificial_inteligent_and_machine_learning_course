```python
data_no_nan.to_csv('Income by state.csv',index=False)
```

```python
state_max = data_no_nan.groupby(['state','date'], as_index = False)['income'].max()
state_max
```

```python
vis_state_max = px.bar(state_max, x='state',y='income',color='date',title='Max income according to each country')

vis_state_max.show()
```

```python
state_min = data_no_nan.groupby(['state','date'], as_index = False)['income'].min()
state_min
```

```python
vis_state_min = px.bar(state_min, x='state',y='income',color='date',title='Min income according to each country')

vis_state_min.show()
```

```python
state_median = data_no_nan.groupby(['state','date'],as_index = False)['income'].median()
state_median
```

```python
vis_state_median = px.bar(state_median, x='state',y='income',color='date',title='Median income according to each country')

vis_state_median.show()
```

```python
state_mean = data_no_nan.groupby(['state','date'], as_index = False)['income'].mean().round(2)
state_mean 
```

```python
vis_state_mean = px.bar(state_mean, x='state',y='income',color='date',title='Mean income according to each country')

vis_state_mean.show()
```
