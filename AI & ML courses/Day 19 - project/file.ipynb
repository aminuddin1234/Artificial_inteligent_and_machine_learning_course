```python
import pandas as pd
import numpy as np
import random

# Setting seed for reproducibility
random.seed(42)
np.random.seed(42)

# Lists for data simulation
allergies_list = ["Gluten", "Shellfish", "Peanuts", "Dairy", "Soy", "None", "Gluten, Dairy"]
conditions_list = ["Diabetes", "Hypertension", "Celiac Disease", "None", "Diabetes, Hypertension"]
goals_list = ["Weight Loss", "Muscle Gain", "Maintenance", "Low Sodium"]
restaurants = ["Green Leaf Cafe", "Seafood Shack", "The Burger Joint", "Pasta Paradise", "Asian Fusion", "Healthy Bites", "The Steakhouse"]
menu_items = {
    "Green Leaf Cafe": ["Quinoa Salad", "Avocado Toast", "Green Smoothie"],
    "Seafood Shack": ["Grilled Salmon", "Lobster Roll", "Shrimp Scampi"],
    "The Burger Joint": ["Classic Cheeseburger", "Veggie Burger", "Bacon Burger"],
    "Pasta Paradise": ["Spaghetti Bolognese", "Fettuccine Alfredo", "Pesto Penne"],
    "Asian Fusion": ["Sushi Platter", "Pad Thai", "Kung Pao Chicken"],
    "Healthy Bites": ["Grilled Chicken Bowl", "Tofu Stir-fry", "Zucchini Noodles"],
    "The Steakhouse": ["Ribeye Steak", "Mashed Potatoes", "Grilled Asparagus"]
}
reviews = [
    "Great meal, very safe for my allergy!",
    "They say it's GF but there was cross-contamination.",
    "Nutritional info was accurate.",
    "The staff was very knowledgeable about ingredients.",
    "Felt sick after eating, might have hidden shellfish.",
    "Perfect for my diet goals.",
    "Menu was confusing regarding allergens.",
    "Highly recommend for diabetics.",
    "The ingredients didn't match the description.",
    "Safe and delicious!"
]

data = []

for i in range(1, 101):
    restaurant = random.choice(restaurants)
    item = random.choice(menu_items[restaurant])
    
    row = {
        "User_ID": f"USR_{1000 + i}",
        "Allergy_Triggers": random.choice(allergies_list),
        "Chronic_Conditions": random.choice(conditions_list),
        "Dietary_Goals": random.choice(goals_list),
        "Restaurant_Name": restaurant,
        "Menu_Item": item,
        "Ingredients_List": "Simulated list of ingredients for " + item,
        "Calories": random.randint(200, 1200),
        "Macros (Protein/Carbs/Fat)": f"{random.randint(10,50)}g/{random.randint(20,100)}g/{random.randint(5,40)}g",
        "Recall_Status": "Safe" if random.random() > 0.05 else "Recall Alert",
        "User_Review_Feedback": random.choice(reviews),
        "Image_Reference": f"img_{i:03d}.jpg",
        "OCR_Menu_Extracted_Status": "Success" if random.random() > 0.1 else "Failed"
    }
    data.append(row)

df = pd.DataFrame(data)

# Save to Excel
output_file = "Palate_AI_Dataset_v2.xlsx"
df.to_excel(output_file, index=False)
print(output_file)
```
