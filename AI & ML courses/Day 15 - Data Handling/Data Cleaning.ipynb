- outliers in the data, we can also use median to fill up NAN data 
- mean is really affected to the outliers 

```python
newdf = df.copy()
mean_price = newdf['Price'].mean().round(2)
print(mean_price)
newdf['Price'].fillna(mean_price,inplace=True)
newdf
```

```python
squared = newdf['Quantity'] ** 2
newquantity = squared / (squared**0.5)
newquantity

newdf['Quantity'] = newquantity
newdf
```

```python
mean_price = newdf['Quantity'].mean().round(2)
print(mean_price)
newdf['Quantity'].fillna(mean_price,inplace=True)
newdf

median_qnty = newdf['Quantity'].median()

newdf['Quantity'].fillna(median_qnty,inplace=True)
newdf
```
