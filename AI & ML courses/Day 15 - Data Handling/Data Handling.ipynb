```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Example dataset
data2 = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, 200, 40],  # 200 is an outlier
}
 
df = pd.DataFrame(data2)
df
```

```python

import seaborn as sns
```

```python
sns.boxplot(data=df)
```

```python
#std ddev for the age
#all points
# now we have formula too
# all data points will have a z score
# if the z score > 3 or <-3 you can treat them as outliers and drop them
```

```python
from scipy import stats
```

```python
z_scores = stats.zscore(df['Age'])
z_scores
```

```python
threshold = 1
# -3 < z <3
new_x = df[ (-threshold < z_scores) & (z_scores < threshold) ]
new_x
```

```python
sns.boxplot(new_x['Age'])
```
