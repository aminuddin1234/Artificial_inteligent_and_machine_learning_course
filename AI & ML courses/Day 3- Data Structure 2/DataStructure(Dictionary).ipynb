#Dictionary is mutable collection of key value pairs.

* Keys are unique
* We denote the dictionary by {key:value}


```python
players={'name': 'Sachin','age':48,'country':'India','is_retired':True}
```

```python
print(players)
```

```python
print('All Keys')
for k in players.keys():
  print(k)
```

```python
print('All Values')
for v in players.values():
  print(v)
```

```python
print('Key -> Value \n')    #\n = create new line
for k in players.keys():
  print(k,'->',players[k])
```

```python
print('is retired?',players['is_retired'])
```

```python
print('Name of Player',players['name'])
```

```python
for k,v in players.items():
  print(k,'\t',v)
```

```python
for item in players.items():
  print(item)
```

```python
players['sport']='Cricket'
```

```python
print(players)
```

```python
student={}
```

```python
while True:
  print('\n Student Record System')
  print('1. Add Student')
  print('2. Display Student')
  print('3. Update Student')
  print('4. Delete Student')
  print('5. Exit')
  choice=int(input('Enter your choice(1-5):\t'))
  if choice==1:
    print('Add Student')
    student['name']=input('Enter student name:\t')
    student['class']=input('Enter student class:\t')
    student['marks']=int(input('Enter student marks:\t'))
    print('Student added successfully')

  elif choice==2:
    if student:
      print('Student Record as follows')
      for k,v in student.items():
        print(k, '\t',v)
      else:
        print('\n No Record found')

  elif choice==3:
    if student:
      key_to_update=input('Enter key to update:(name,class,marks)\t')
      if key_to_update in student.keys():
        new_value=input('Enter new value:\t')
        student[key_to_update]=new_value
        print('Record updated successfully')
      else:
        print('Invalid key')

    else:
        print('\n No Record found')

  elif choice==4:
     student.clear()
     print('Record deleted successfully')


  elif choice==5:
          print('Good Bye')
          break
  else:
          print('Invalid choice (choose 1-5)')
```
