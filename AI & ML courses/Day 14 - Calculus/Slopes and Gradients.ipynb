# Gradient descant

is the process where the algorithm does when to find global minimum in error to cut cost

```python
x = np.linspace(-5,5)
 
y = 2 * (x**2) + 3
 
result = np.array(y)

plt.plot(x,result)
```

```python
x = np.linspace(-5,5)
 
y = -2 * (x**2) + 3 # when negative the line will be terbalik
 
result = np.array(y)

plt.plot(x,result)
```

```python
x = np.linspace(-10,10)
 
y = 2 * ((x+3)**2) + 3
 
result = np.array(y)

plt.plot(x,result)
plt.axhline(0, color = 'green', linewidth = 0.6)
plt.axvline(0, color = 'purple', linewidth = 0.6)
plt.show()
```

```python
x = np.linspace(-10,10)
 
y = (3* (x**3)) + (2* (x **2)) + ((x - 1) **2) + 4
 
result = np.array(y)

plt.plot(x,result)
plt.axhline(0, color = 'green', linewidth = 0.6)
plt.axvline(0, color = 'purple', linewidth = 0.6)
plt.show()
```

```python
x = np.linspace(-10,10,10000)
y = (x)**4 - 80*(x**2) + 50*x
result = np.array(y)

plt.plot(x,result)
```

- based on this graph we don't know the global maximum where it's, we just know the local maximum has 3 ... and based on the graph also we also know global minimum that is between -5.0 to -7.5
