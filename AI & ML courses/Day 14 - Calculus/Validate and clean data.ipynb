```python
industry_list = ['Artificial intelligence', 'Other','E-commerce & direct-to-consumer', 'Fintech',\
       'Internet software & services','Supply chain, logistics, & delivery', 'Consumer & retail',\
       'Data management & analytics', 'Edtech', 'Health', 'Hardware','Auto & transportation', \
        'Travel', 'Cybersecurity','Mobile & telecommunications']
```

```python
set(modi_comp['Industry']) - set(industry_list)
```

```python
replacement_dict = {'Artificial Intelligence': 'Artificial intelligence',
                   'Data management and analytics': 'Data management & analytics',
                   'FinTech': 'Fintech'
                   }

modi_comp['Industry'] = modi_comp['Industry'].replace(replacement_dict)

set(modi_comp['Industry']) - set(industry_list)
```

```python
modi_comp[modi_comp.duplicated(subset=['Company'], keep=False)]
```

```python
companies = modi_comp.drop_duplicates(subset=['Company'], keep='first')
```

```python
modi_comp['High Valuation'] = pd.qcut(modi_comp['Valuation'], 2, labels = ['low', 'high'])
```

```python
modi_comp['Continent'].value_counts()
```

```python
continent_dict = {'North America': 1,
                  'Asia': 2,
                  'Europe': 3,
                  'South America': 4,
                  'Oceania': 5,
                  'Africa': 6
                 }
modi_comp['Continent Number'] = modi_comp['Continent'].replace(continent_dict)
modi_comp.head()
```

```python
modi_comp['Country/Region Numeric'] = modi_comp['Country/Region'].astype('category').cat.codes
```

```python
industry_encoded = pd.get_dummies(modi_comp['Industry'])

modi_comp = pd.concat([modi_comp, industry_encoded], axis=1)
```
