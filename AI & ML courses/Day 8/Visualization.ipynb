```python
import matplotlib.pyplot as plt
#Create a bar plot
plt.bar(students,grades, color = 'green') #specify properties of our bar chart
plt.title('Chart of student grades')
plt.xlabel('student names')
plt.ylabel('marks')
plt.show() #incase you need it for another platform
```

```python

plt.scatter(students,grades, color = 'green')
plt.title('Chart of student grades')
plt.xlabel('student names')
plt.ylabel('marks')
plt.show() 
```

```python
#Create a line chart
plt.plot(students,grades, color = 'green') 
plt.title('Chart of student grades')
plt.xlabel('student names')
plt.ylabel('marks')
plt.show() 
```

```python
#All of them in the same place
#create a figure of 1 row and 3 column
plt.figure(figsize= (15,5))

#sub plot
plt.subplot(1,3,1) #rows, column position
plt.bar (students,grades, color = 'green')

plt.subplot(1,3,2)
plt.plot (students,grades, color = 'green')

plt.subplot(1,3,3)
plt.scatter (students,grades, color = 'green')
```

```python
cities = ['KL','SG','NY','Mex','Chile']
temperature = [30,33,26,34,20]
```

```python
plt.figure(figsize= (15,5))

#sub plot
plt.subplot(1,3,1) #rows, column position
plt.bar (cities,temperature, color = 'yellow')
plt.title('Chart of cities temperature')
plt.xlabel('Cities name')
plt.ylabel('Temperature in ')

plt.subplot(1,3,2)
plt.plot (cities,temperature, color = 'red')
plt.title('Chart of cities temperature')
plt.xlabel('Cities name')
plt.ylabel('Temperature in ')

plt.subplot(1,3,3)
plt.scatter (cities,temperature, color = 'black')
plt.title('Chart of cities temperature')
plt.xlabel('Cities name')
plt.ylabel('Temperature in ')
```
