**Tuple: is an ordered immutable collection of items**
* we use() to denote tuple

```python
numbers=(100,130,100,230,400,500,100)
```

```python
print('Number of items in tuple are: ',len(numbers))
```

```python
print('All items in tuple')
for items in numbers:
  print(items,end= '\t')
```

```python
for (index,num) in enumerate(numbers):
  print (index,num)
```

**set** âœˆ
* set is an unordered collection of unique **(duplicate not allow)** items, its mutable,
* we denote set by {}

```python
our_first_set={100,200,300,400,500}
print(our_first_set)
```

```python
second_set={'ac','cooler','fan','washing machine','ac','fan','fridge'}
print(second_set)
```

```python
print('Number of items in second set : ',len(second_set))
```

```python
second_set.remove('fan')
print('Set after removing fan: Number of items are: ',len(second_set))
print(second_set)
```

```python
second_set.add('mobile')
print('Set after add mobile: Number of items are: ',len(second_set))
print(second_set)
```

```python
second_set.clear()
print('Set after clearing all items: Number of items are: ',len(second_set))
print(second_set)
```

```python
set_one={10,20,30,40,50}
set_two={5,10,15,20,25,30,35,40}
```

```python
print('Number of items in set_one',len(set_one))
print('Number of items in set_two',len(set_two))
common=set_one.intersection(set_two)
print('Common items in both sets: ',common)
```

```python
all_items=set_one.union(set_two) # .union segala benda yang x duplicate
print('All items in both sets: ',all_items)
```

```python
diff_items=set_two.difference(set_one) # .difference ... benda yang x sama or x berulang
print('items in set_one but not in set_two',diff_items)
```

```python
course_two={'Python','Excel','SQL','Soft Skill','Azure','GitHub','Tensorflow'}
```

```python
print('Number of items in set_one',len(course_one))
print('Number of items in set_two',len(course_two))
common=course_one.intersection(course_two)
print('Common items in both sets: ',common)
```
