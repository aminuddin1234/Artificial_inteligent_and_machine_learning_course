```python
class BankAccount:
  def __init__(self,account_number,account_holder,balance):
    self.account_number=account_number
    self.account_holder=account_holder
    self.balance=balance
    print('Account Created')

  def display_balance(self):
    print('Account Balance:',self.balance)

  def deposit(self,amount):
    self.balance+=amount
    print('Amount Deposited')
    print('Current Balance:',self.balance)

  def withdraw(self,amount):
    if self.balance>=amount:
      self.balance-=amount
      print('Amount Withdrawn')
      print('Current Balance:',self.balance)
    else:
      print('Insufficient Balance')

  def __str__(self):
    return f'Account Number: {self.account_number} \t Account Holder: {self.account_holder} \t Balance: {self.balance}'
```

```python
account1=BankAccount('ACC-1001','Amin',1000000)
```

```python
account2=BankAccount('ACC-1675','Wani',20000)
```

```python
account3=BankAccount('ACC-1344','Pija',30000)
```

```python
print('Account 2 Details')
print(account2)
```

```python
account3.display_balance()    #objectname.methodname
```
