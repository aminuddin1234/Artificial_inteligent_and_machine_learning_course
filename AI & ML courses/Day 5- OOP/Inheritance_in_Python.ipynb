```python
class Employee:
  def __init__(self,id,name,salary):
    self.id=id
    self.name=name
    self.salary=salary
  def __str__(self):
    return f'Id={self.id},name={self.name},salary={self.salary}'
```

```python
emp=Employee(1,'Sam',5600)
print(emp)
```

```python
class Developer(Employee):
  def __init__(self, id, name, salary,domain,project):
    super().__init__(id, name, salary)
    self.domain=domain
    self.project=project

  def __str__(self):
    return f'Id={self.id},name={self.name},salary={self.salary},domain={self.domain},project={self.project}'
```

```python
dev=Developer(14,'Aminuddin',7000,'Python','Project 3')
print(dev)
```

```python
#Example 3
class Product:
  def reg(self):
    self.pid=int(input('Enter Product Id:'))
    self.pname=input('Enter Product Name:')
    self.price=float(input('Enter Product Price:'))

  def display(self):
    print('product id \t',self.pid)
    print('product name \t',self.pname)
    print('product price \t',self.price)
```

```python
class HerbalProduct(Product):
  def reg(self):
    super().reg()
    self.herbs=input('Enter Herbal Ingredients: \t')
    self.mfdate=input('Enter Product mfdate: \t')
    self.expdate=input('Enter Product expdate: \t')

  def display(self):
    super().display()
    print('Herbal Ingredients used \t',self.herbs)
    print('Product mfdate \t',self.mfdate)
    print('Product expdate \t',self.expdate)
```

```python
hp1=HerbalProduct()
hp1.reg()
hp1.display()
```
