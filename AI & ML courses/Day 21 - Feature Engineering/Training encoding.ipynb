```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.info()
```

```python
titanic.isnull().sum()
```

```python
titanic['class'].unique() # ['First', 'Second', 'Third']
titanic['sex'].unique()  # ['male', 'female']
titanic['alive'].unique() # ['no', 'yes']
```

```python
titanic_codes = {
    'First': 0,
    'Second': 1,
    'Third': 2
}
titanic['map_class_codes'] = titanic['class'].map(titanic_codes)
titanic
```

```python
titanic_codes = {
    'male': 0,
    'female':1
}
titanic['map_sex_codes'] = titanic['sex'].map(titanic_codes)
titanic
```

```python
titanic_codes = {
    'no': 0,
    'yes':1
}
titanic['map_alive_codes'] = titanic['alive'].map(titanic_codes)
titanic
```

```python
x,y = pd.factorize(titanic['class'])
titanic['factorize_class'] = x
```

```python
x,y = pd.factorize(titanic['sex'])
titanic['factorize_sex'] = x
```

```python
x,y = pd.factorize(titanic['alive'])
titanic['factorize_alive'] = x
titanic
```

```python
titanic['class'] = titanic['class'].astype('category')
titanic['class_cat_codes'] = titanic['class'].cat.codes
```

```python
titanic['sex'] = titanic['sex'].astype('category')
titanic['sex_cat_codes'] = titanic['sex'].cat.codes
```

```python
titanic['alive'] = titanic['alive'].astype('category')
titanic['alive_cat_codes'] = titanic['alive'].cat.codes
titanic
```

```python
from sklearn.preprocessing import LabelEncoder
titanic['class_L_E'] = LabelEncoder().fit_transform(titanic['class'])
```

```python
titanic['sex_L_E'] = LabelEncoder().fit_transform(titanic['sex'])
```

```python
titanic['alive_L_E'] = LabelEncoder().fit_transform(titanic['alive'])
```

```python
titanic
```

# One Hot Encoder Training

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
titanic_training = sns.load_dataset('titanic')
```

```python
onehot_titanic = pd.get_dummies(titanic_training,columns=['class'])
onehot_titanic['class_First'].dtype
onehot_titanic
```

```python
onehot_titanic = pd.get_dummies(titanic_training,columns=['sex'])
onehot_titanic['sex_male'].dtype
onehot_titanic
```

```python
encoder = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
titanic_class = ['class']
```

```python
titanic_array_codes  = encoder.fit_transform(onehot_titanic[titanic_class])
titanic_array_codes
```

```python
titanic_code = encoder.get_feature_names_out(titanic_class)
titanic_code
```

```python
onehot_titanic[titanic_code] = titanic_array_codes

```

```python
onehot_titanic
```

```python
onehot_titanic.drop(columns =['pclass'],inplace =True)
```

```python
onehot_titanic
```

```python
titanic_alive = ['alive']
titanic_coded_alive = encoder.fit_transform(onehot_titanic[titanic_alive])
```

```python
alive_names_out = encoder.get_feature_names_out(titanic_alive)
```

```python
onehot_titanic[alive_names_out] = titanic_coded_alive
onehot_titanic.drop(columns=['alive'],inplace=True)
```

```python
onehot_titanic
```

```python
titanic_2 = sns.load_dataset('titanic')
```

```python
titanic_columns = (['alive', 'survived'])
```

```python
array_titanic = encoder.fit_transform(titanic_2[titanic_columns])
array_titanic
```

```python
code_titanic = encoder.get_feature_names_out(titanic_columns)
```

```python
titanic_2[code_titanic] = array_titanic
```
