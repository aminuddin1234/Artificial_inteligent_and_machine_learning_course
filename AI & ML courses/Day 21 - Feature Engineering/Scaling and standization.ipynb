```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python
data = pd.DataFrame({
    'age': [20,50,100],
    'income': [20000,50000,100000]
})

scaler = MinMaxScaler()

normalize_data = scaler.fit_transform(data)
df_normal = pd.DataFrame(normalize_data,columns=['age','income'])
df_normal
```

- Why do we normalize data
    - easier to calculate
    - 
- a lot of preferences to salary over age in the example
- our model only use number

- 100-20 = 80
- 50-20 = 30
- 30/80 = 0.375
