```python
# assignment operator =e.g. username=sam1234
# equal == e.g. if(number==5)
atm_pin=7654
allowed_attemps=3
no_attempts=0

while(allowed_attemps!=0):
  user_pin=int(input('Enter your pin:\t'))
  if(user_pin==atm_pin):
    print('Welcome to ATM')
    break

  else:
    allowed_attemps-=1
    no_attempts+=1

    if(allowed_attemps==0):
      print('Your card has been blocked')
    else:
      print('Wrong Pin Try Again')

```

**exercise**

write a number game to find out hidden number between (1 to 10).

*   ask the user to guess the hidden number
*   note: hidden number is 7
*   if he entered number less than 7 show too low
*   if he entered number more than 7 show too high
*   if he entered correct number show correct answer
    You have won and end the loop
    only 3 attempts are allowed
    if he have not guess within 3 attempts show you have lost your attempts







```python
hidden_number=7
allowed_attemps=3
no_attempts=0

while(allowed_attemps!=0):
  number=int(input('Enter hidden number:\t'))
  if(number==hidden_number):
    print('You have won and end the loop')
    break

  else:
    allowed_attemps-=1
    no_attempts+=1

  if(allowed_attemps==0):
    print('you have lost your attempts')
  if(number<7):
    print('too low')
  if(number>7):
    print('too high')
```
