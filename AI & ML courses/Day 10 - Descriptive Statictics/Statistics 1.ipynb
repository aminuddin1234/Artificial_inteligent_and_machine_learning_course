Median  = 
1. sort the numbers
2. targetted the middle number

```python
nums = [20, 5, 50,60, 130, 80]
nums.sort()
print(nums)

mean = sum(nums) /len(nums)
print('Mean of the list: \t',mean)

#if the length is odd
median = nums[len(nums)//2]
print(median)

#if length is even
median= (nums[len(nums)//2] + nums[(len(nums)//2) + 1]) /2
print(median)
```

Explore how to sort a list manually

* create a program, that takes a list, and returns a sorted list
* only allowed to google, when you have struggled at least 1 hours

```python
mylist = [64, 34, 25, 12, 22, 11, 90, 5]

n = len(mylist)
for i in range(n-1):
  for j in range(n-i-1):
    if mylist[j] > mylist[j+1]:
      mylist[j], mylist[j+1] = mylist[j+1], mylist[j]

print(mylist)
```

```python
nums = [20, 5, 60,100, 130, 80]
nums.sort()
print(nums)

mean = sum(nums) /len(nums)
print('Mean of the list: \t',mean)

if len(nums)%2 ==0:
#if the length is odd
    median= (nums[len(nums)//2] + nums[(len(nums)//2) -1]) /2
    print('Median of the: \t', median)

else:
#if length is even
    median = nums[len(nums)//2]
    print('Median of the: \t', median)
```
