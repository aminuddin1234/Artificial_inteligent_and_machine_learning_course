* Recursive function: if the function is calling to itself, it's known as recursive function image.png

```python
print('find out factorial of a number')
def factorial(num):
  if(num==1):
    return 1
  else:
    return num*factorial(num-1)
```

```python
print(factorial(1))
```

```python
print(factorial(2))
```

```python
print(factorial(3))
```

```python
print(factorial(5))
#5 x 4 x 3 x 2 x 1
```

```python
print(factorial(8))
# 8 x 7 x 6 x 5 x 4 x 3 x 2 x 1
```

```python
#Recursive function example -2
def reverse_string(text):
  if len(text) == 0:
    return ""
  else:
    return (reverse_string(text[1:])+text[0])
```

```python
print('reverse of "hello" is: \t',reverse_string("hello"))
```

```python
user_text=input('Enter text to reverse:  ')
print('reverse of',user_text,'is',reverse_string(user_text))
```

```python
list_birds= ['pigeon','parrot','sparrow','penguin','peacock']
```

```python
list_birds.sort(reverse=True)
list_birds=' #'.join(list_birds)

print('list of birds (decending): ',list_birds)
```
