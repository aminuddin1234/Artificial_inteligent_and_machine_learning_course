```python
#Variable Length Arguments
def addition(*args):
  total=0
  print('\n All Numbers are: ')
  for i in args:
    print(i)
    total+=i
  return total
```

```python
print('\n Result: ',addition(12,45))
print('\n Result: ',addition(12,45,55,89))
print('\n Result: ',addition(12,45,44,23,78,90))
```

Write a function to take a number from user and print the cube and square of given number
* cube of number means number * number * number
* e.g. cube of 2 is 8 and 3 is 27
* square of number means number*number
*e.g square of 2 is 4 & 3 is 9


```python
def cube(n1,n2,n3):
  return n1*n2*n3
```

```python
def square(n1,n2):
  return n1*n2
```

```python
num=int(input('Enter your Number:\t'))

print(f'cube of {num} is: ',cube(num,num,num))
```

```python
num2=int(input('Enter your Number:\t'))

print(f'square of {num2} is: ',square(num2,num2))
```

```python
def cube_square(num):
  print('Cube of',num, 'is', num*num*num)
  print(f'Cube of{num} is {num,num,num}')
  print(f'Square of {num} is {num*num}')
```

```python
given_number=int(input('Enter a number: \t'))
cube_square(given_number)
```

```python
#if C convert celcius to Farenheit
#if F convert Farenheit to celcius

def convert_temp(temp,unit='C'):
  if unit=='C':
    return ((temp*9/5)+32)
  elif unit=='F':
    return ((temp-32)*5/9)
  else:
    return 'Invalid Unit , Please use only C or F'
```
