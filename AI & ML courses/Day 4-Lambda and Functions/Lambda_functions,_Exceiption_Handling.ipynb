```python
addition=lambda num1,num2:num1+num2

multiplication=lambda num1,num2:num1*num2

division=lambda num1,num2:num1/num2

subtraction=lambda num1,num2:num1-num2

average=lambda num1,num2:(num1+num2)/2
```

```python
fnum=float(input('Enter first number:\t'))
snum=float(input('Enter second number:\t'))
print(f'Addition of {fnum} and {snum} is {addition(fnum,snum)}')
print(f'Multiplication of {fnum} and {snum} is {multiplication(fnum,snum)}')
print(f'Division of {fnum} and {snum} is {division(fnum,snum)}')
print(f'subtraction of {fnum} and {snum} is {subtraction(fnum,snum)}')
print(f'Average of {fnum} and {snum} is {average(fnum,snum)}')
```

```python
def div(num1,num2):
  result=num1/num2
  return result
```

```python
try:
  num1=int(input('Enter first number: \t'))
  num2=int(input('Enter second number: \t'))
  print(f'Result after dividing {num1} and {num2} is: ',div(num1,num2))
except ZeroDivisionError as ze:
  print('Error Occured!!! Division by zero is not possible, Please put again',ze)
  num2=int(input('Enter second number: \t'))
  print(f'Result after dividing {num1} and {num2} is: ',div(num1,num2))

except Exception as ex:
  print('Error Occured!!! ',ex)
finally:
  print('End Of program')
```

```python
###########
def to_int(value):
  if value is None or value == "":
    return "Not Given"
  try:
    return int(value)
  except (ValueError,TypeError):
    return "Conversion Failed"

```
